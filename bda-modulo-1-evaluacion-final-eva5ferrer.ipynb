### 1. agregar_producto(nombre, precio, cantidad) : Agrega un producto al inventario o actualiza su cantidad si ya existe. Debe recibir el nombre, precio y cantidad del producto como parámetros.

```python
# RECORDATORIO ESTRUCTURA: inventario = [{nombre: aaa, precio: bbb, cantidad: ccc}, {nombre: ddd, precio: eee, cantidad: fff}, etc..]
inventario = []
```

```python
def agregar_producto (nombre, precio, cantidad):
    for producto in inventario: #aqui recorremos toda la lista con los diferentes productos (diccionarios):
        if producto["nombre"] == nombre:
            producto["cantidad"] += cantidad 
            print(f"Este producto ya existe, así que hemos actualizado su cantidad.")
            return # si después de recorrer todo el inventario, encuentra el producto, actualiza la cantidad, ejecuta el return y la función se para.
    
    # el append solamente se ejecuta si el if no se cumple. Así que podemos actualizar el inventario con el nuevo producto:
    inventario.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})
    print("Este producto no lo teníamos en el inventario, así que lo hemos añadido correctamente")
    
```

```python
#Hago varios ejemplos para comprobar que la función se ejecuta correctamente:
agregar_producto("camiseta", 15, 2)
agregar_producto("falda", 18, 4)
agregar_producto("sombrero", 32, 2)
agregar_producto("falda", 18, 3) #aquí se nos tendría que actualizar la cantidad de faldas que tenemos en el inventario.
```

```python
def actualizar_stock(nombre, cantidad):
    for producto in inventario:
        if producto["nombre"].lower() == nombre.lower():
            producto["cantidad"] += cantidad
            print(f"Hemos actualizado la cantidad de producto en el inventario.")
            return 
    print(f"El producto no está en el inventario, así que no hemos actualizado su cantidad.")
           
    
```

### 5. eliminar_producto(nombre): Elimina un producto del inventario por nombre. Debe recibir el nombre del producto como parámetro.

* Busca el producto por nombre.
* Elimina el producto del inventario si existe.

```python
inventario 
```

```python
def eliminar_producto(nombre):
    for producto in inventario:
        if producto["nombre"].lower() == nombre.lower():
            inventario.remove(producto)
            print("Hemos eliminado el producto del inventario")
            return
    print("El producto no está en el inventario, así que no se puede eliminar.")
```

### 6. calcular_valor_inventario(): Calcula y muestra el valor total del inventario. 
Utiliza un bucle for para calcular el valor total del inventario.
Itera a través del inventario y suma el valor de cada producto (precio x cantidad). Es decir, calcula el valor total del inventario.

```python
def calcular_valor_inventario():
    valor_de_cada_producto = 0.0 #creo una variable nueva para añadir la suma de cada producto.
    # la creo fuera del bucle for, porque sino se estaría creando una nueva en cada vuelta.
    for producto in inventario:
        valor_de_cada_producto += producto["precio"] * producto["cantidad"] 
    print(f"El valor total del inventario es de {valor_de_cada_producto} euros.")
    return 
```

```python
calcular_valor_inventario()
```

### 7. realizar_compra(): Permite a un cliente realizar una compra seleccionando productos del inventario. Debe interactuar con el cliente para seleccionar productos y calcular el costo total de la compra. 
* Utiliza un bucle while para permitir al cliente realizar múltiples compras.
* Muestra el inventario y solicita al cliente ingresar el nombre del producto que desea comprar.
* Registra los productos seleccionados en un carrito y actualiza el inventario.
* Calcula el costo total de la compra.

```python
inventario
```

```python
def realizar_compra():
    carrito_compra = []
    costo_total_compra = 0.0
    ver_inventario()

    while True:
        producto_escogido = input("""Te he listado los productos que tenemos disponible abajo, escribe el nombre del producto que quieres comprar. 
                                  Cuando ya no quieras comprar más, escribe STOP.""").lower()
        if producto_escogido.upper() == "STOP": 
            print("Entendido, ya no quieres comprar más.")
            break
        producto_encontrado = False #aqui asumo que el producto no está en el inventario   
        for producto in inventario: #empieza a buscar en el inventario
            if producto_escogido.lower() == producto["nombre"].lower(): 
                producto_encontrado = True #Lo cambio a true, cuando el producto está en el inventario

                if producto["cantidad"] > 0:
                    carrito_compra.append(producto_escogido) #solamente podemos añadir el producto en el carrito, si tenemos almenos uno de cantidad.    
                    print(f"Has elegido: {producto_escogido}.")     
                    costo_total_compra += producto["precio"]    
                    producto["cantidad"] -=1     
                    print(f"Tu carrito de la compra incluye los siguientes productos: {carrito_compra}, que suman un coste total de {costo_total_compra} euros.")    
                else: #cuando la cantidad es 0 o menor:
                    print(f"Lo sentimos, no tenemos suficiente cantidad de producto en el inventario.")
                break

        if producto_encontrado == False:
            print(f"Lo sentimos, actualmente no tenemos este producto disponible.")
            
    return print(f"Este es el inventario actual de la tienda: {inventario}.") #para ver si se actualiza el inventario 
           
            
```
